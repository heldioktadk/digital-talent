```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
run("show tables")
```

```python
run("SELECT country,state,city FROM offices")
```

# 9. What is the minimum payment received?

```python
run("""
SELECT 
    YEAR(paymentDate) AS Tahun,
    MONTH(paymentDate) AS Bulan,
    MIN(amount) AS PembayaranTermurah 
FROM payments 
GROUP BY Tahun,Bulan
ORDER BY PembayaranTermurah ASC
LIMIT 0,1
""")
```

# 10. List all payments greater than twice the average payment.

```python
run("""
SELECT 
   *
from payments
where amount > ((SELECT avg(amount)*2 FROM payments)) 
""")
```

# 11. What is the average percentage markup of the MSRP on buyPrice?

```python
run("SELECT MSRP,buyPrice,(MSRP - buyPrice) AS untung, ((MSRP - buyPrice)/2) AS Persen  FROM products HAVING untung > buyPrice")
```

# 13. Report the name and city of customers who don't have sales representatives?

# 14. What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.


```python
run("""
SELECT CONCAT(firstName, " ", lastName) AS fullName, jobTitle
FROM employees
WHERE jobTitle LIKE '%VP%' OR jobTitle LIKE '%Manager%'
""")
```

# 15. Which orders have a value greater than USD 5,000?

```python
run("""
SELECT orderNumber, SUM(priceEach * quantityOrdered) AS Value
FROM orderdetails
GROUP BY orderNumber
HAVING Value > 5000
""")
```
