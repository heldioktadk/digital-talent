# Deskripsi Data
   Data ini berisi mengenai jumlah siswa SMK yang putus sekolah per Provinsi dari tahun 2008-2011. dalam dataset ini berisikan variabel:
   1. kode_provinsi: Kode Provinsi
   2. nama_provinsi: Nama Provinsi
   3. tahun: Tahun
   4. jumlah_putus_sekolah: Jumlah siswa SMK yang putus sekolah
   5. persen_putus_sekolah: Persentase siswa SMK yang putus sekolah

```python
import pandas as pd
data_putus_sekolah = pd.read_csv('data-putus-sekolah-smk-2008-2011.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
data_putus_sekolah
```

# Statistik Tentang Data

```python
data_putus_sekolah.jumlah_putus_sekolah.count()
```

```python
stps = data_putus_sekolah.jumlah_putus_sekolah.sum()
minps = data_putus_sekolah.jumlah_putus_sekolah.min()
maxps = data_putus_sekolah.jumlah_putus_sekolah.max()
print("Jumlah Anak yang Putus Sekolah :", stps)
print("Jumlah Anak Yang Putus Sekolah Terkecil :", minps)
print("Jumlah Anak Yang Putus Sekolah Terbanyak :", maxps)
```

```python
data_putus_sekolah.groupby('nama_provinsi').mean()
```

# Visualisasi Sederhana Jumlah Anak yang Putus Sekolah SMK Per Provinsi

```python
ax = data_putus_sekolah.groupby('nama_provinsi').mean()['jumlah_putus_sekolah']
ax = ax.plot(kind='bar', title ="Jumlah Anak yang Putus Sekolah Berdasarkan Provinsi", figsize=(15,15), legend=True, fontsize=12)
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Jumlah Anak Yang Putus Sekolah", fontsize=12)
```

# Visualisasi Sederhana Jumlah Anak yang Putus Sekolah SMK Per Tahun

```python
ax1 = data_putus_sekolah.groupby('tahun').mean()['jumlah_putus_sekolah']
ax1 = ax1.plot(kind='bar', title ="Jumlah Anak yang Putus Sekolah Berdasarkan Tahun", figsize=(15,15), legend=True, fontsize=12)
ax1.set_xlabel("Tahun", fontsize=12)
ax1.set_ylabel("Jumlah Anak Yang Putus Sekolah", fontsize=12)
```
